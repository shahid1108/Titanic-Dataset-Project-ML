 # Key Observation of Scatterplot :

- Passengers of all age groups are present in both survival categories (0 = did not survive, 1 = survived).

- There is no strong linear relationship between age and survival — both younger and older passengers are distributed across survival outcomes.

- However, it appears that more younger passengers survived, as there is slightly denser clustering of points at lower ages near Survived = 1.

- Some very young and very old individuals survived, suggesting age alone was not a decisive factor, though there may be patterns when combined with    other features like class or sex.

```python
sns.scatterplot(data=data,x="Survived",y="Age");
```

# Key Observation of Scatterplot :

- Females (y = female) have more points near Survived = 1, indicating a higher survival rate.

- Males (y = male) are more concentrated near Survived = 0, indicating a lower survival rate.

- This supports the well-known pattern from the Titanic dataset: females were more likely to survive than males, likely due to the "women and children first" evacuation strategy.

```python
sns.scatterplot(data=data,x="Survived",y="Sex");
```

# Key Observation for the jointplot:

- Survivors (Survived = 1) are concentrated more in the younger age range (around 20–40 years), with a visible peak around ages 20–30.

- Non-survivors (Survived = 0) show a wider spread, including a significant number of individuals in the 20–40 age range as well, but with fewer from the youngest age group.

- The histograms show that most passengers were between 20 and 40 years old.

- A small number of children (under 10) and elderly (above 60) are present in both categories, but survival was more common among younger passengers overall.

```python
sns.jointplot(data=data,x="Survived",y="Age",kind="hist");
```

# Observation for the pivot table:

- Females had a higher survival rate across almost all age groups where data is available, with many age groups showing a survival rate of 1.0 (100%).

- Males show much lower survival rates, especially between the ages of 10 to 50, where the values are mostly below 0.5.

- For young children (age ~0.42 to 4 years), both males and females had very high survival rates, showing that younger children were prioritized for rescue.

- Older age groups (60+) show very few data points, but survival was still possible, especially for females (e.g., survival = 1.0 at age 62, 63 for females).

- NaN values indicate missing data or absence of passengers in that specific sex-age combination.

```python

pd.pivot_table(data, values='Survived', index='Sex', columns='Age', aggfunc='mean')
```

# key Observation of histplot:

- A higher number of passengers did not survive the Titanic disaster, as shown by the tall bar near 0 (over 500).

- The bar near 1 (survived) is shorter, indicating fewer passengers survived (around 340).

- The histogram confirms that the dataset is imbalanced, with more non-survivors than survivors.

-  The KDE curve (Kernel Density Estimate) gives a smoothed view and supports the same pattern — two clear peaks around 0 and 1, showing the binary nature of survival.

```python
sns.histplot(data['Survived'], bins=20, kde=True)
plt.title('Total Survived ')
plt.xlabel('Survived')
plt.ylabel('Frequency')
plt.show()
```

# Key Observation of Boxplot:

-  The majority of members' ages are clustered between roughly 20 and 40 years old. That big blue box in the middle really highlights this central tendency.

-  The median age seems to be right around 30. That line inside the blue box gives us a good sense of the middle point.

-  There are definitely some older members. The whisker extending to the right shows ages going up to around 60.

- And a few potential outliers on the older side. Those circles beyond the right whisker indicate ages that are quite a bit higher than the main group.

- The younger ages are less spread out than the older ones. The left whisker is shorter than the right one.

- No members appear to be younger than around 0. The left whisker starts close to zero.

```python
sns.boxplot(x=data['Age'])
plt.title('Age of Member')
plt.show()
```

# Observation of the Scatterplot :

- The data points are clustered vertically at the "Survived" values of 0 and 1.

- This vertical clustering indicates that "Survived" is likely a categorical variable.

-  A value of 0 likely represents passengers who did not survive.

-  A value of 1 likely represents passengers who survived.

-  The ages of the passengers are distributed across both survival categories (0 and 1).

-  There isn't an obvious linear relationship or trend visible between age and survival based on this plot alone.

```python
sns.scatterplot(x='Survived', y='Age', data=data)
plt.title('Survived vs Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```
